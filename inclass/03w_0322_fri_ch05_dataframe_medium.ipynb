## 05-3 파생변수 만들기 : 시험예상!!!!!!!

```python
df = pd.DataFrame({"var1" : [4,3,8],
                   "var2" : [2,6,1]})
df
```

```python
df1 = df
df1
```

```python
df1["var_sum"] = df1["var1"] + df1["var2"] #파생 변수 만들기
df1
```

```python
df # df에도 반영이 됨 -> .copy() 하면 반영 안됨
```

```python
# 열 varmean <- var1과 var2의 평균
df["var_mean"] = (df1["var1"] + df1["var2"]) / 2
df1
```

```python
df2 = df.copy()
df2
```

```python
df2.sum(axis = 1)
```

```python
df2["my_sum"] = df2.sum(axis=1)
df2
```

```python
# 열 my_mean <- var1과 var2의 평균
df2["my_mean"] = df2[["var1","var2"]].mean(axis=1) #axis=1 가로로 구해라  , 시험!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
df2
```

## mpg 통합 연비 변수 만들기

```python
# mpg 데이터 불러오기
mpg = pd.read_csv("mpg.csv")
mpg.head()
```

```python
# 통합 연비 변수(total) 만들기: total <= cty, hwy의 평균 시험!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

mpg["total"] = (mpg["cty"] + mpg["hwy"]) /2 #구식 방법 -> 괄호 넣어야 됨.....
mpg.head()
```

```python
mpg["mpg_mean"] = mpg[["cty", "hwy"]].mean(axis='columns') #axis='columns' or 1
mpg.head()
```

```python
sum(mpg["total"]) / len(mpg) # total 합계를 행 수로 나누기
```

```python
mpg['total'].mean() # 통합 연비 변수 평균
```
