### 함수와 메서드 차이 알아보기

1. 내장 함수 : 가장 기본적인 함수 형태, 함수 이름/괄호를 입력하여 사용
 - 파이썬에 내장되어 있어서 패키지 설치나 불러오기 없음

2. 패키지 함수 : 패키지 이름을 먼저 입력한 다음, 점을 찍고 함수 이름과 괄호를 입력하여 사용
 - 패키지 로드 필요

3. 메서드(method)(=기술) : 변수가 지니고 있는 함수, 변수명 입력한 다음 점을 찍고 메서드 이름과 괄호를 입력하여 사용
 - 모든 변수에서 사용 가능한 것은 아님
 - 변수의 자료 구조에 따라 메서드가 달라짐
 - type()을 통해 자료 구조 파악 가능
 - ex) df.head() -> 가능, val.head() -> 불가

4. 어트리뷰트(능력치) 알아보기
  - attribute  : 변수가 지니고 있는 값
  - 변수명 뒤에 점을 찍고 어트리뷰트 이름을 입력하여 사용
  - 메서드처럼 변수명 뒤에 점을 찍지만 괄호는 입력하지 않음 -> 괄호가 x 어트리뷰트
  - df.head() -> 메서드, df.shape -> 어트리뷰트
  - 변수의 자료구조에 따라 달라짐
  - ex) df.shape -> 가능, val.shape -> 불가

### mpg 통합 연비 변수 만들기

### 조건문을 활용해 파생변수 만들기
- 함수를 이용해 파생변수를 생성 가능
- 조건에 따라 다른 값을 반환하는 조건문 함수로 파생변수 생성
- ex) 자동차 연비 기준을 충족해 고연비 합격 판정을 받은 자동차가 몇 대나 되는지 알아보고 기준값을 넘으면 합격, 아니면 불합격을 부여하기

```python
# 1. 기준값 정하기 -> 몇을 기준을 합격 여부를 판단할지 정하기
mpg['total'].describe() # 통합 연비 변수의 요약 통계량 출력
```

```python
# 히스토그램을 통해 연비 분포 알아보기
mpg['total'].plot.hist()
```

```python
# 2. 합격 판정 변수 만들기 -> NumPy 패키지의 where()이용하면 조건에 따라 서로 다른 값을 부여
import numpy as np # numpy : 배열 연산, 통계치 계산 등 수치 연산을 할 때 사용하는 패키지
```

```python
# 막대 그래프로 빈도 표현
count_test = mpg['test'].value_counts()
```

```python
count_test.plot.bar()
```

### 중첩 조건문 활용하기
- A,B,C 세 종류의 연비 등급으로 분류한 변수 만들기
- total이 30이상이면 A, 20~29는 B, 20미만 C등급으로 분류
- 범주 값을 3가지 이상 부여하기 위해 np.where()안에 다시 np.where()을 넣는 형식 사용 -> 조건문을 여러 개 중첩한 중첩 조건문


```python
# 1. 연비 등급 변수 만들기
mpg['grade'] = np.where(mpg['total']>=30, 'A', np.where(mpg['total']>=20, 'B', 'C'))
mpg.head()
```

```python
# 2. 빈도표와 막대 그래프로 연비 등급 확인
count_grade = mpg['grade'].value_counts()
count_grade
```

```python
count_grade.plot.bar(rot=0)
```

```python
# 알파벳순으로 막대 정렬 -> 빈도 기준으로 내림차순으로 정렬되어 있는 것을 알파벳순으로 정렬
count_grade = mpg['grade'].value_counts().sort_index()
count_grade
```

```python
count_grade.plot.bar(rot=0)
```

```python
# 필요한 만큼 범주 만들기 -> 범주의 수 -1 만큼 np.where()을 중첩하기
# A,B,C,D 등급 변수 만들기
mpg['grade2'] = np.where(mpg['total']>=30, 'A',
                         np.where(mpg['total']>=25, 'B',
                                  np.where(mpg['total']>=20, 'C', 'D')))
mpg.head()
```

### 목록에 해당하는 행으로 변수 만들기
- 조건 중 하나에 해당하면 특정 값을 부여해 파생변수 만들 때가 있음
- ex) mpg 데이터의 category가 compact, subcompact, 2seater면 small 부여하고 그렇지 않으면 large를 부여한 파생변수를 만들기
- 또는을 의미하는 |(파이프 기호) 사용
- mp.where()에 여러 조건을 입력할 때는 각 조건에 괄호 입력해야 함

```python
mpg['size'] = np.where((mpg['category']=='compact')|(mpg['category']=='subcompact')|(mpg['category']=='2seater'), 'small', 'large')
mpg['size'].value_counts()
```

```python
# df.isin() 사용하여 코드를 간편하게 작성 ->[]이용해 조건 목록으로 입력 , 변수의 값이 입력한 목록에 해당하는지 확인하는 기능
mpg['size']  = np.where(mpg['category'].isin(['compact', 'subcompact', '2seater']), 'small', 'large')
mpg['size'].value_counts()
```
