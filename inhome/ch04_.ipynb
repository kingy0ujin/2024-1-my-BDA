# 04-2 데이터 프레임 만들기 - 시험 성적 데이터를 만들어 보자!

### 데이터 입력해 데이터 프레임 만들기

- DataFrame({'변수(컬럼)명':[변수에 넣을 값],[],[]})


```python
import pandas as pd
# pandas : 데이터를 가공할 때 사용
```

```python
df = pd.DataFrame({'name': ['김지훈', '이유진','박동현','김민지'],
                   'english' : [90, 80, 60, 70],
                   'math' : [50, 60, 100, 20]})
df
```

### 데이터 프레임으로 분석하기

- 특정 변수(컬럼)의 값 추출하기 -> 데이터프레임 이름['변수(컬럼)명']

```python
df['english']
```

- 변수의 값으로 합계 구하기

```python
sum(df['english']) # 영어 점수의 합계
```

```python
sum(df['math']) # 수학 점수의 합계
```

- 변수의 값으로 평균 구하기

```python
sum(df['english'])/4 # 영어 점수 평균
```

```python
sum(df['math'])/4 # 수학 점수 평균
```

- 혼자서 해보기

```python
# Q1
sales = pd.DataFrame({'제품' : ['사과','딸기','수박'],
                   '가격' : [1800, 1500, 3000],
                   '판매량' : [24, 38, 13]})
sales
```

```python
# Q2
print(sum(sales['가격']) /3)
print(sum(sales['판매량']/3))
```

### 데이터 프레임을 csv 파일로 저장하기
