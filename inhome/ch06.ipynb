## 06-5 파생변수 추가하기
- 변수를 조합, 함수를 이용해 새 변수를 만들어 분석할 수 있음
- df.assign()을 이용 -> 파생변수 생성

### 파생변수 추가하기
- exam.assign(새 변수명 = 변수를 만드는 공식)
- 변수명은 자유, 따옴표 입력 x

```python
# total
exam.assign(total = exam['math']+exam['english']+exam['science'])
```

```python
# 여러 파생변수 한 번에 추가하기
# 쉼표 -> 새 변수명, 변수 만드는 공식을 나열

exam.assign(total = exam['math'] + exam['english'] + exam['science'] ,
            mean = (exam['math'] + exam['english'] + exam['science']) / 3)
```

### df.assign()에 np.where()' 적용하기

```python
import numpy as np
exam.assign(test = np.where(exam['science'] >= 60, 'pass', 'fail'))
```

### 추가한 변수를 pandas 함수에 바로 활용하기
- 변수를 추가하면 이어지는 판다스 함수에 바로 활용 가능

```python
# total 변수 추가, total 기준 정렬
exam.assign(total = exam['math'] + exam['english'] + exam['science']) \
 .sort_values('total')
```

### lambda 이용해 데이터 프레임명 줄여 쓰기
- df.assign() 사용시 변수명 앞에 데이터 프레임명을 반복 입력 -> lambda 로 데이터 프레임명 대신 약어 사용
- lambda x: -> 데이터 프레임명 자리에 x를 입력하겠다는 의미

```python
# 긴 데이터 프레임명 지정
long_name = pd.read_csv('exam.csv')

# long_name 직접 입력
long_name.assign(new = long_name['math'] + long_name['english'] + long_name['science'])

# long_name 대신 x 입력
long_name.assign(new = lambda x: x['math'] + x['english'] + x['science'])
```

```python
# 앞에서 만든 변수를 활용해 다시 변수 만들기 -> lambda를 이용해 데이터프레임을 약어로 입력해야 함
# 여기서 앞이란 같은 셀 기준? 앞 셀에서 만든 파생 변수를 활용하니 에러가 뜸
exam.assign(total = exam['math'] + exam['english'] + exam['science'],
            mean = lambda x : x['total'] / 3)
```

```python
# 파생변수를 만드는 행에도 lambda 적용하기
exam.assign(total = lambda x : x['math'] + x['english'] + x['science'],
            mean = lambda x : x['total'] / 3)
```

## 06-6 집단별로 요약하기
- 집단별 평균, 집단별 빈도 -> 각 집단을 요약한 값을 구할 때
- df.groupby(), df.agg() 함수 사용 -> 집단 간에 어떤 차이가 있는지 파악 가능

### 집단별로 요약하기
- 전체 요약 통계량 구하기
- 요약 통계량 -> df.agg()
- df.agg(요약값을 할당할 변수명 = (값을 요약하는데 사용할 변수와 함수 나열))

```python
# math 평균 구하기
exam.agg(mean_math = ('math', 'mean')) # 변수명에는 따옴표 x, 요약하는 사용할 변수에는 따옴표 o
```

### 집단별로 요약 통계량 구하기
- 전체 평균은 .mean() 사용
- df.agg()는 전체를 요약이 아니라 df.groupby()에 적용해 집단별 요약값을 구할 때 사용
- df.groupby()에 변수 지정 -> 변수의 범주별로 데이터를 분리 -> df.agg()적용 -> 집단별 요약 통계량 구함

```python
# 반별 수학 점수 평균
exam.groupby('nclass').\
  agg(mean_math = ('math','mean'))
```

```python
# groupby()의 기본값이 변수를 인덱스로 바꾸도록 설정되어 있음
# df.groupby(as_index = False) -> 변수를 원래대로 유지
# 인덱스 : 값이 데이터 프레임의 어디에 있는지 위치를 나타낸 값

exam.groupby('nclass', as_index = False)\
  .agg(mean_math = ('math','mean'))
```

### 여러 요약 통계량 한 번에 구하기
- df.agg -> 한 번에 구하기 가능

```python
exam.groupby('nclass') \
  .agg(mean_math = ('math', 'mean'),
       sum_math = ('math','sum'),
       median_math = ('math','median'),
       n = ('nclass', 'count')) #빈도(반 별 학생수)
```

### 집단별로 다시 집단 나누기
- 성적 데이터를 반별로 나눈 다음 다시 성별로 나눠 각 반의 성별 평균 점수를 구할 수 있음


```python
# mpg -> 제조 회사별로 집단을 나눈 다음 다시 구동 방식별로 나눠 도시 연비 평균을 구하기
mpg.groupby(['manufacturer', 'drv']) \
  .agg(mean_cty = ('cty', 'mean'))
```

```python
# audi의 drv별 빈도
mpg.query('manufacturer == "audi"') \
  .groupby(['drv'])\
  .agg(n = ('drv', 'count'))
```

```python
# value_counts()로 집단별 빈도 간단하게 구하고 내림차순으로 정렬 / 시리즈 자료구조 / query()적용 x
mpg['drv'].value_counts()
```

```python
# query()를 적용하기 위해 데이터프레임으로 변환하기
mpg['drv'].value_counts()\
  .to_frame('n')\
  .rename_axis('drv')\
  .query('n > 100')
```

### pandas 함수 조합하기
- 절차가 복잡해 보이는 분석도 조합으로 간단하게 해결 가능


```python
# 제조 화사별로 suv 자동차의 도시 및 고속도로 합산 연비 평균을 구해 내림차순으로 정렬 후 5위까지 출력
mpg.query('category == "suv"')\
  .assign(total = (mpg['cty'] + mpg['hwy']) / 2)\
  .groupby('manufacturer')\
  .agg(mean_tot = ('total', 'mean'))\
  .sort_values('mean_tot', ascending=False)\
  .head()
```

```python
# 어떤 회사에서 compact 차종을 가장 많이 생산하는지 알아보려고 합니다,
# 회사별 compact 차종 수를 내림차순으로 정렬해 출력하세요.
mpg.query('category == "compact"')\
  .groupby('manufacturer')\
  .agg(n = ('category', 'count'))\
  .sort_values('n', ascending=False)
```

## 06-7 데이터 합치기
- 가로로 합치기 : 기존 데이터에 변수(열)을 추가
- 세로로 합치기 : 기존 데이터에 행을 추가

### 가로로 합치기
- pd.merge() 이용
- 결합할 데이터 프레임명 나열
- 오른쪽에 입력한 데이터 프레임을 왼쪽 데이터 프레임에 결합 -> how = 'left'
- 데이터를 합칠 때 기준으로 삼을 변수 -> on에 입력

```python
# 중간고사 데이터 만들기
test1 = pd.DataFrame({'id' : [1,2,3,4,5],
                     'midterm' : [60, 80, 70, 90, 85]})

# 기말고사 데이터 만들기
test2 = pd.DataFrame({'id' : [1,2,3,4,5],
                      'final' : [70, 83, 65, 95, 80]})
```

```python
test1
```

```python
test2
```

```python
#id를 기준으로 합쳐서 total 할당
total = pd.merge(test1, test2, how = 'left', on = 'id')
total
```

### 세로로 합치기
- pd.concat() 이용
- 결합할 데이터 프레임 명을 []를 이용해 나열
- 세로로 합칠 때는 두 데이터의 변수명이 동일해야 함
- 변수명이 다를 시 pd.rename() 이용

```python
# 학생 1~5번 시험 데이터 만들기
group_a = pd.DataFrame({'id' : [1,2,3,4,5],
                        'test' : [60, 80, 70, 90, 85]})

# 학생 6~10번 시험 데이터 만들기
group_b = pd.DataFrame({'id' : [6,7,8,9,10],
                        'test' : [70, 83, 65, 90, 80]})
```

```python
group_a
```

```python
group_b
```

```python
# 데이터 합쳐서 group_all에 할당
group_all = pd.concat([group_a, group_b])
group_all
```
